#SUBPLOTTING

```python
#plotting for showing two graphs in one graph
x = [1,2,3]
y = [20,50,40]
x1 = [6,9,3]
y2 = [8,1,4]
plt.plot(x,y,label = "first",linewidth = 3)
plt.plot(x1,y2, label = "second")
#this function is used to draw labels on the top of the graph
plt.legend()
plt.show()
```

```python
#plotting for showing two graphs in one graph
x = [1,2,3]
y = [20,50,40]
x1 = [6,9,3]
y2 = [8,1,4]
#this function is used to draw histogram
plt.hist(x,label = "first",linewidth = 3)
plt.plot(x1,y2, label = "second")
#this function is used to draw labels on the top of the graph
plt.legend()
plt.show()
```

```python
#plotting for showing two graphs in one graph
x = [1,2,3]
y = [20,50,40]
x1 = [6,9,3]
y2 = [8,1,4]
plt.scatter(x,y,label = "first",linewidth = 3)
plt.plot(x1,y2, label = "second")
#this function is used to draw labels on the top of the graph
plt.legend()
plt.show()
```

#BAR PLOT OF CATEGORICAL DATA

```python
#program for simple bar plot of categorical data
names = ['akshat','anant','abhishek']
values = [90,30,20]
plt.bar(names,values)
plt.show()
```

```python
#program for histogram plot of categorical data
names = ['akshat','anant','abhishek']
plt.hist(names)
plt.show()
```

#HANDLING TEXT

```python
values = [90,30,20]
plt.hist(values,label = "first")
plt.xlabel("data")
plt.ylabel("num")
plt.legend()
#this syntax is used to add text in the plot
plt.text(40,0.4, ' $\mu = 100$')
plt.show()
```

text annotations

```python
fig , geeeks = plt.subplots()
  
t = np.arange(0.0, 5.0, 0.001)
s = np.cos(3 * np.pi * t)
line = geeeks.plot(t, s, lw = 2)
  
# Annotation
geeeks.annotate('Local Max', xy =(3.3, 1),xytext =(3, 1.8), arrowprops = dict(facecolor ='green',shrink = 0.05),)
  
geeeks.set_ylim(-2, 2)
  
# Plot the Annotation in the graph
plt.show()
```
